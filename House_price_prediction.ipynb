## Feature Engineering.

```python
df1
```

```python
df1['Area'] = np.log(df1['Area'])
```

```python
df1['BHK_OR_RK'].replace( {'BHK':1, 'RK': 0}, inplace = True)
```

```python
df1['POSTED_BY'] = df1['POSTED_BY'].astype('uint8')
df1['UNDER_CONSTRUCTION'] = df1['UNDER_CONSTRUCTION'].astype('bool')
df1['RERA'] = df1['RERA'].astype('bool')
df1['BHK_NO.'] = df1['BHK_NO.'].astype('int8')
df1['BHK_OR_RK'] = df1['BHK_OR_RK'].astype('bool')
df1['READY_TO_MOVE'] = df1['READY_TO_MOVE'].astype('bool')
df1['RESALE'] = df1['RESALE'].astype('bool')
```

```python
df1['POSTED_BY'].describe()
```

```python
df1.dtypes
```

```python
df1['City'] = df1['ADDRESS'].apply(lambda x: x.split(',')[-1])
df1['Address'] = df1['ADDRESS'].apply(lambda x: x.split(',')[0])
df1.drop('ADDRESS', axis = 1, inplace=True)
```

#### Extracting new features

```python
df1['median_sqft_per_bhkno'] = df1.groupby('BHK_NO.')['Area'].transform('median')
df1['min_sqft_per_bhkno'] = df1.groupby('BHK_NO.')['Area'].transform('min')
df1['max_sqft_per_bhkno'] = df1.groupby('BHK_NO.')['Area'].transform('max')

df1['mean_sqft_per_location'] = df1.groupby('LONGITUDE')['Area'].transform('median')
df1['min_sqft_per_location'] = df1.groupby('LONGITUDE')['Area'].transform('min')
df1['max_sqft_per_location'] = df1.groupby('LONGITUDE')['Area'].transform('max')

df1['median_sqft_per_city'] = df1.groupby('City')['Area'].transform('median')
df1['min_sqft_per_city'] = df1.groupby('City')['Area'].transform('min')
df1['max_sqft_per_city'] = df1.groupby('City')['Area'].transform('max')

df1['median_sqft_per_address'] = df1.groupby('Address')['Area'].transform('median')
df1['min_sqft_per_address'] = df1.groupby('Address')['Area'].transform('min')
df1['max_sqft_per_address'] = df1.groupby('Address')['Area'].transform('max')
```

```python
df1.head()
```

```python
df1['Area_per_room'] = df1.apply(lambda x: x['Area']/x['BHK_NO.'], axis = 1)
```

```python
df1
```

```python
# apply le on categorical feature columns
df1[categorical_cols] = df1[categorical_cols].apply(lambda col: le.fit_transform(col))
```

## Modelling

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
df1.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
y =df1.Transaction_Price
x= df1.drop('Transaction_Price',axis =1)
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2,random_state =1234)
```

```python
print(len(x_train),len(x_test),len(y_train),len(y_test))
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
pipelines = {
    'lasso' : make_pipeline(StandardScaler(),Lasso(random_state =123)),
    'ridge' : make_pipeline(StandardScaler(),Ridge(random_state =123)),
    'enet'  : make_pipeline(StandardScaler(),ElasticNet(random_state =123)),
    'rf'    : make_pipeline(StandardScaler(),RandomForestRegressor(random_state =123)),
    'gb'    : make_pipeline(StandardScaler(),GradientBoostingRegressor(random_state =123))
}
```

```python
len(pipelines)
```

```python
for key,value in pipelines.items():
    print(key,type(value))
```

```python
pipelines['lasso'].get_params()
```

```python
lasso_hyperparams = {
    'lasso__alpha': [0.001,0.005,0.01,0.05,0.1,0.5,1,5,10]
}

ridge_hyperparams = {
    'ridge__alpha': [0.001,0.005,0.1,0.5,0.01,0.05,1,5,10]
}

enet_hyperparams = {
    'elasticnet__alpha' :[0.001,0.005,0.01,0.05,0.1,0.5,1,5,10],
    'elasticnet__l1_ratio' : [0.1,0.3,0.5,0.7,0.9]
}
rf_hyperparams = {
    'randomforestregressor__n_estimators' : [100,200],
    'randomforestregressor__max_features' : ['auto','sqrt',0.33]
}
gb_hyperparams ={
    'gradientboostingregressor__n_estimators': [100,200],
    'gradientboostingregressor__learning_rate' : [0.05,0.1,0.2],
    'gradientboostingregressor__max_depth' : [1,3,5]
}
```

```python
hyperparams = {
    'rf' : rf_hyperparams,
    'gb' : gb_hyperparams,
    'lasso' : lasso_hyperparams,
    'ridge' : ridge_hyperparams,
    'enet' : enet_hyperparams
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
fitted_models={}
for name,pipeline in pipelines.items():
    model = GridSearchCV(pipeline,hyperparams[name],cv =10,n_jobs =-1)
    model.fit(x_train,y_train)
    fitted_models[name]  = model
    print(name,'has been fitted.')
```

```python
for key,value in fitted_models.items():
    print(key,type(value))
```

```python
for name,model in fitted_models.items():
    print(name,model.best_score_)
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print('R^2:',r2_score(y_test,pred))
print('MAE:',mean_absolute_error(y_test,pred))
```
